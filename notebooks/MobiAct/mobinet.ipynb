```python
import pandas as pd
from glob import glob
```

```python
path = '../data/MobiAct/Raw Data/'
```

```python
activities = ['STD', 'WAL', 'JOG', 'JUM', 'STU', 'STN', 'SCH', 'SIT', 'CHU', 'CSI', 'CSO', 'LYI']
```

```python
#glob(path+activities[0]+'/*_acc*')
```

```python
data = []
for ac in activities:
    for f in glob(path+ac+'/*_acc*'):
        df = pd.read_csv(f, comment='#').reset_index()
        df['Activity'] = ac
        df['User'] = f.split('_')[2]
        data.append(pd.DataFrame(df))
```

```python
df = pd.concat(data)
df
```

```python
df = df.rename(columns={'level_0': 'Time', 'level_1': 'x', 'level_2': 'y', '@DATA': 'z'})
```

```python
df.to_csv('../data/MobiAct/raw_data.csv', index=False)
```
